```python
from __future__ import print_function
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize, StandardScaler
from sklearn.utils.class_weight import compute_sample_weight
from sklearn.metrics import roc_auc_score
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
raw_data = pd.read_csv('creditcardfraud/creditcard.csv')
print("There are " + str(len(raw_data)) + " observations in the CC fraud data.")
print("There are " + str(len(raw_data.columns)) + " variables in the dataset.")

raw_data.head()
raw_data.dtypes

```

```python
n_replicas = 10

# inflating dataset
big_raw_data = pd.DataFrame(np.repeat(raw_data.values, n_replicas, axis=0), columns = raw_data.columns)
print("There are " + str(len(big_raw_data)) + " observations in the CC fraud data.")
print("There are " + str(len(big_raw_data.columns)) + " variables in the dataset.")
big_raw_data.head()
```

```python
labels = big_raw_data.Class.unique()
print(labels)

sizes = big_raw_data.Class.value_counts().values
print(sizes)

fig, ax = plt.subplots()
print(fig)
print(ax)
ax.pie(sizes, labels=labels, autopct='%1.3f%%')


ax.set_title('Target Variable Value Counts')
plt.show()
big_raw_data
```

```python
plt.hist(big_raw_data.Amount.values, 6, histtype='bar', facecolor='g')
plt.show()

print('min amount value is ', np.min(big_raw_data.Amount.values))
print('max amount value is ', np.max(big_raw_data.Amount.values))
print('90% of the transaction have an amount less or equal than ', np.percentile(raw_data.Amount.values, 90))
big_raw_data.head()
```

<h1> Data Preprocessing </h1>

```python
big_raw_data.iloc[:, 1:30] = StandardScaler().fit_transform(big_raw_data.iloc[:, 1:30])
data_matrix = big_raw_data.values
# big_raw_data.iloc[:, 1:30].head()
# data_matrix
X = data_matrix[:, 1:30]

y = data_matrix[:, 30]
X = normalize(X, norm="l1")
print('X.shape=', X.shape, 'y.shape=', y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42, stratify = y)
print('X_train.shape=', X_train.shape, 'Y_train.shape=', y_train.shape)
print('X_test.shape=', X_test.shape, 'Y_test.shape=', y_test.shape)
```

```python
w_train = compute_sample_weight('balanced', y_train)
from sklearn.tree import DecisionTreeClassifier
```

```python
sklearn_dt = DecisionTreeClassifier(max_depth=4, random_state=35)

t0 = time.time()
sklearn_dt.fit(X_train, y_train, sample_weight = w_train )
sklearn_time = time.time()- t0
print("[Scikit-Learn] Training time (s): {0: .5f}".format(sklearn_time))
```

```python
from snapml import DecisionTreeClassifier
```

```python
snapml_dt = DecisionTreeClassifier(max_depth = 4, random_state = 45, n_jobs = 4)

t0 = time.time()
snapml_dt.fit(X_train, y_train, sample_weight = w_train)
snapml_time = time.time() - t0
print("[Snam ML] Training time (s): {0: 5f}".format(snapml_time))
```

```python
# snapml_dt = DecisionTreeClassifier(max_depth = 4, random_state = 45, n_jobs = 4, use_gpu = True)

# t0 = time.time()
# snapml_dt.fit(X_train, y_train, sample_weight = w_train)
# snapml_time = time.time() - t0
# print("[Snam ML] Training time GPU (s): {0: 5f}".format(snapml_time))
```

```python
training_speedup = sklearn_time/snapml_time
print('[Decision Tree Classifier] Snap ML vs. Scikit-Learn speedup : {0: 2}x '.format(training_speedup))

```

```python
sklearn_pred = sklearn_dt.predict_proba(X_test)[:,1]
sklearn_pred
```

```python
# Receiver Operating Characteristices, area Under Receiver - True Positives, False negatives
sklearn_roc_auc = roc_auc_score(y_test, sklearn_pred)
print('[Scikit-Learn] ROC-AUC score: {0: .3f}'.format(sklearn_roc_auc))
```

```python
snapml_pred = snapml_dt.predict_proba(X_test)[:,1]
snapml_pred
```

```python
snapml_roc_auc = roc_auc_score(y_test, snapml_pred)
print('[Snap ML] ROC-AUC score: {0: .3f}'.format(snapml_roc_auc))
```
